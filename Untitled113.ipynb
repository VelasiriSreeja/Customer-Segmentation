```python
m=df.drop_duplicates(inplace=False)
m
```

```python
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df['TotalSpend'] = df['Quantity'] * df['UnitPrice']
```

```python
df = df[['CustomerID', 'InvoiceNo', 'InvoiceDate', 'Quantity', 'UnitPrice', 'Country', 'TotalSpend']]
```

```python
print("Cleaned Data Shape:", df.shape)
```

```python
latest_date = df['InvoiceDate'].max()
```

```python
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (latest_date - x.max()).days,  # Recency
    'InvoiceNo': 'nunique',                                 # Frequency
    'TotalSpend': 'sum'                                     # Monetary
}).reset_index()
```

```python
rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']
```

```python
rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1]).astype(int)
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5]).astype(int)
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5]).astype(int)
```

```python
rfm['RFM_Score'] = rfm['R_Score'] + rfm['F_Score'] + rfm['M_Score']
```

```python
rfm_for_clustering = rfm[['Recency', 'Frequency', 'Monetary']]
```

```python
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_for_clustering)
```

```python
inertia = []
K = range(2, 10)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(rfm_scaled)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(7,5))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
optimal_k = 4
```

```python
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
rfm['Cluster'] = kmeans.fit_predict(rfm_scaled)
```

```python
sil_score = silhouette_score(rfm_scaled, rfm['Cluster'])
print(f"Silhouette Score: {sil_score:.3f}")
```

```python
cluster_summary = rfm.groupby('Cluster').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': ['mean', 'count']
}).round(2)
print(cluster_summary)
```
